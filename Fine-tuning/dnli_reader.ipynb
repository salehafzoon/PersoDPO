```python
def repair_train_data():
    path = './dnli/dnli/dialogue_nli_extra/dialogue_nli_EXTRA_uu_train.jsonl'
    with open(path) as f:
        txt = f.readline()
        t = txt[:-44]
        t = t + ']'
        t = t.strip("'<>() ").replace('\'', '\"')
        n = json.dumps(t)
        json_obj = json.loads(n)
        with open('./dnli/repaired_train.json', 'w') as fout:
            json.dump(json_obj, fout)
        
```

```python
repair_train_data()
```
